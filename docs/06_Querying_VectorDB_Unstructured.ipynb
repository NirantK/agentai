```python
@tool(registry=vector_db_registry)
def query_vector_db(query: Query):
    """
    Ask the vector database a question
    """
    print(f"Querying vector database: {query}")
    results = client_db.get_docs(query=query)
    return results


question = f"""Search for the Deep Learning Based Document Image Analysis in the vector database.
    Get me three results from the vector database and include the documents and distances."""

conversation = Conversation()
conversation.add_message(
    "user",
    question,
)

chat_response = chat_complete_execute_fn(conversation, tool_registry=vector_db_registry, model="gpt-3.5-turbo")
print(chat_response)
```
